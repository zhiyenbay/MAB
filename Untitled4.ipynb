```python
%matplotlib inline
import random
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
# set up pandas table display
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
import seaborn as sns
from __future__ import division
```

```python
def initialize(n_arms):
    counts = [0 for col in range(n_arms)]
    values = [0.0 for col in range(n_arms)]
    return counts, values
```

```python
def ind_max(x):
    m = max(x)
    return x.index(m)
```

```python
def select_arm(epsilon):
    if random.random() > epsilon:
        return ind_max(values)
    else:
        return random.randrange(len(values))
```

```python
def update(chosen_arm, reward):
    counts[chosen_arm] = counts[chosen_arm] + 1
    n = counts[chosen_arm]

    value = values[chosen_arm]
    new_value = ((n - 1) / n) * value + (1 / n) * reward
    values[chosen_arm] = new_value
```

```python
def average(values):
    result = 0.0
    for value in values:
        result = result + value
    return result / len(values)
```

```python
n = len(items.keys())
print n
```

```python
epsilon = 0.5
x = []
y = []
counts, values = initialize(n)
for i in range(0,500):
    random_arm = select_arm(epsilon)
    x.append(i)
    print i
    if A[random_arm][0] < len(A[random_arm])-1:
        A[random_arm][0] = A[random_arm][0] + 1
        current_reward = A[random_arm][ A[random_arm][0] ]
        print current_reward, random_arm
        update(random_arm, int(current_reward))
        print "maximum value: " + str(max(values))
    y.append(max(values))
```
